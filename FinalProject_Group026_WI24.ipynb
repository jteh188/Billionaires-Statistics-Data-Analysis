## Create Final Dataset

```python
#Create new column in main dataframe for high_income countries
high_i = []
for c in billionaires_dropped['country']:
    if c in high_income_countries:
        high_i.append(True)
    else:
        high_i.append(False)
billionaires_dropped['high_income'] = high_i
```

```python
#Convert GDP to int value, return if NaN
def convert_GDP(string):
    if string != string:
        return string
    else:
        string = string.lower()
        string = string.strip()
        string = string.replace("$","")
        string = string.replace(",","")
        return int(string)
    

billionaires_dropped['gdp'] = billionaires_dropped['gdp_country'].apply(convert_GDP)
```

```python
# import functions for statistical tests
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
```

```python
# mean age of each gender
male_age = billionaires_dropped[billionaires_dropped['gender'] == 'M']['age']
mean_age_male = male_age.mean()
print(mean_age_male)

female_age = billionaires_dropped[billionaires_dropped['gender'] == 'F']['age']
mean_age_female = female_age.mean()
print(mean_age_female, '\n')

# t-test of differences in ages between genders
t_statistic, p_value = ttest_ind(male_age, female_age)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
# mean age of high income vs. not high income countries

age_high_income = billionaires_dropped[billionaires_dropped['high_income'] == True]['age']
mean_age_high_income = age_high_income.mean()
print(mean_age_high_income)

age_low_income = billionaires_dropped[billionaires_dropped['high_income'] == False]['age']
mean_age_low_income = age_low_income.mean()
print(mean_age_low_income, '\n')

# t-test of differences in ages between high-income and not high-income countries
t_statistic, p_value = ttest_ind(age_high_income, age_low_income)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

There is no significant difference among the ages of male versus female billionaires, as well as no significant difference among the ages of billionaires from high income vs. not high income countries. The p-values for each t-test are above 0.05, suggesting that the differences between groups are due to chance.

```python
# mean age of gender + high income breakdowns

male_age_high_income = billionaires_dropped[(billionaires_dropped['gender'] == 'M') & (billionaires_dropped['high_income'] == True)]['age']
female_age_high_income = billionaires_dropped[(billionaires_dropped['gender'] == 'F') & (billionaires_dropped['high_income'] == True)]['age']
male_age_low_income = billionaires_dropped[(billionaires_dropped['gender'] == 'M') & (billionaires_dropped['high_income'] == False)]['age']
female_age_low_income = billionaires_dropped[(billionaires_dropped['gender'] == 'F') & (billionaires_dropped['high_income'] == False)]['age']

mean_age_male_high_income = male_age_high_income.mean()
mean_age_female_high_income = female_age_high_income.mean()
mean_age_male_low_income = male_age_low_income.mean()
mean_age_female_low_income = female_age_low_income.mean()

print(mean_age_male_high_income)
print(mean_age_female_high_income)
print(mean_age_male_low_income)
print(mean_age_female_low_income, '\n')

# F test for gender, age, and high-income vs. not high-income groups
# Is also not significant since t-tests were not significant
data = [male_age_high_income, female_age_high_income, male_age_low_income, female_age_low_income]
f_statistic, p_value = f_oneway(*data)
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

### Tax Rate per Country

Besides, we are also interested in the relationship between the tax rate of a country and the number of billionaires from the country.

```python
tax_rate_dict = {}

for i in range(len(billionaires)):
    if billionaires.loc[i, 'country'] in billionaires['country'].value_counts().index:
        tax_rate_dict[billionaires.loc[i, 'country']] = billionaires.loc[i, 'total_tax_rate_country']

tax_rate = pd.DataFrame(tax_rate_dict.items(), columns=['country', 'tax rate']).merge(
    pd.DataFrame(
        dict(billionaires['country'].value_counts()).items(), 
        columns=['country', 'number of billionaires']
    )[['country', 'number of billionaires']]
)

tax_rate = tax_rate.dropna(subset=['tax rate']).sort_values('number of billionaires', ascending=False)

tax_rate
```
