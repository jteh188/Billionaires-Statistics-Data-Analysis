## Create Final Dataset

```python
#Create new column in main dataframe for high_income countries
high_i = []
for c in billionaires_dropped['country']:
    if c in high_income_countries:
        high_i.append(True)
    else:
        high_i.append(False)
billionaires_dropped['high_income'] = high_i
```

```python
#Convert GDP to int value, return if NaN
def convert_GDP(string):
    if string != string:
        return string
    else:
        string = string.lower()
        string = string.strip()
        string = string.replace("$","")
        string = string.replace(",","")
        return int(string)
    

billionaires_dropped['gdp'] = billionaires_dropped['gdp_country'].apply(convert_GDP)
```

```python
# import functions for statistical tests
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
```

```python
# mean age of each gender
male_age = billionaires_dropped[billionaires_dropped['gender'] == 'M']['age']
mean_age_male = male_age.mean()
print(mean_age_male)

female_age = billionaires_dropped[billionaires_dropped['gender'] == 'F']['age']
mean_age_female = female_age.mean()
print(mean_age_female, '\n')

# t-test of differences in ages between genders
t_statistic, p_value = ttest_ind(male_age, female_age)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
# mean age of high income vs. not high income countries

age_high_income = billionaires_dropped[billionaires_dropped['high_income'] == True]['age']
mean_age_high_income = age_high_income.mean()
print(mean_age_high_income)

age_low_income = billionaires_dropped[billionaires_dropped['high_income'] == False]['age']
mean_age_low_income = age_low_income.mean()
print(mean_age_low_income, '\n')

# t-test of differences in ages between high-income and not high-income countries
t_statistic, p_value = ttest_ind(age_high_income, age_low_income)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

There is no significant difference among the ages of male versus female billionaires, as well as no significant difference among the ages of billionaires from high income vs. not high income countries. The p-values for each t-test are above 0.05, suggesting that the differences between groups are due to chance.

```python
# mean age of gender + high income breakdowns

male_age_high_income = billionaires_dropped[(billionaires_dropped['gender'] == 'M') & (billionaires_dropped['high_income'] == True)]['age']
female_age_high_income = billionaires_dropped[(billionaires_dropped['gender'] == 'F') & (billionaires_dropped['high_income'] == True)]['age']
male_age_low_income = billionaires_dropped[(billionaires_dropped['gender'] == 'M') & (billionaires_dropped['high_income'] == False)]['age']
female_age_low_income = billionaires_dropped[(billionaires_dropped['gender'] == 'F') & (billionaires_dropped['high_income'] == False)]['age']

mean_age_male_high_income = male_age_high_income.mean()
mean_age_female_high_income = female_age_high_income.mean()
mean_age_male_low_income = male_age_low_income.mean()
mean_age_female_low_income = female_age_low_income.mean()

print(mean_age_male_high_income)
print(mean_age_female_high_income)
print(mean_age_male_low_income)
print(mean_age_female_low_income, '\n')

# F test for gender, age, and high-income vs. not high-income groups
# Is also not significant since t-tests were not significant
data = [male_age_high_income, female_age_high_income, male_age_low_income, female_age_low_income]
f_statistic, p_value = f_oneway(*data)
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

### Linear regression between country GDP and number of billionaires

```python
import patsy
import statsmodels.api as sm
```

```python
billionaires_country = billionaires_dropped['country'].value_counts()
billionaires_country
```

```python
test_gdp = gdp_dropped_dup
test_billionaires = billionaires_country
test_gdp_billionaires = test_gdp.merge(test_billionaires, on = 'country', how = 'left', indicator = True)
test_gdp_billionaires['gdp_country'] = test_gdp_billionaires['gdp_country'].replace({'\$':''}, regex = True)
test_gdp_billionaires['gdp_country'] = test_gdp_billionaires['gdp_country'].replace({'\,':''}, regex = True)
test_gdp_billionaires['gdp_country'] = test_gdp_billionaires['gdp_country'].to_numpy().astype(int)
```

```python
outcome, predictors = patsy.dmatrices("count ~ gdp_country", test_gdp_billionaires)
model = sm.OLS(outcome, predictors)
res_log = model.fit()
res_log.summary()
```

Based on the data collected about each the GDP of each country compared to the number of billionaires they have, we ran a linear regression to see if there was a correlation between a country's GDP and the number of billionaires living there. We see a high value for our t-statistic which indicates that the higher the GDP, the more billionaires there are in that country. The p-value also indicates that there is a strong correlation between a countries GDP and the number of billionaires it has.

### Tax Rate per Country

Besides, we are also interested in the relationship between the tax rate of a country and the number of billionaires from the country.

```python
tax_rate_dict = {}

for i in range(len(billionaires)):
    if billionaires.loc[i, 'country'] in billionaires['country'].value_counts().index:
        tax_rate_dict[billionaires.loc[i, 'country']] = billionaires.loc[i, 'total_tax_rate_country']

tax_rate = pd.DataFrame(tax_rate_dict.items(), columns=['country', 'tax rate']).merge(
    pd.DataFrame(
        dict(billionaires['country'].value_counts()).items(), 
        columns=['country', 'number of billionaires']
    )[['country', 'number of billionaires']]
)

tax_rate = tax_rate.dropna(subset=['tax rate']).sort_values('number of billionaires', ascending=False)

tax_rate
```

```python
# drop null values first in order to do t-test
edu_dropped_dup = edu_dropped_dup.dropna()
```

```python
# t-test for primary enrollment between high income vs. low-income countries

high_primary = edu_dropped_dup[edu_dropped_dup['high_income'] == True]['gross_primary_education_enrollment_country']
mean_high_primary = high_primary.mean()
print(mean_high_primary)

low_primary = edu_dropped_dup[edu_dropped_dup['high_income'] == False]['gross_primary_education_enrollment_country']
mean_low_primary = low_primary.mean()
print(mean_low_primary, '\n')

t_statistic, p_value = ttest_ind(high_primary, low_primary)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

This t-test indicates that there is not a significant difference in gross primary enrollment between high-income and low-income countries. The p-value is less than 0.05, so any potential differences observed in primary education between high-income and low-income countries are likely due to chance.

```python
# t-test for tertiary enrollment between high income vs. low-income countries

high_tertiary = edu_dropped_dup[edu_dropped_dup['high_income'] == True]['gross_tertiary_education_enrollment']
mean_high_tertiary = high_tertiary.mean()
print(mean_high_tertiary)

low_tertiary = edu_dropped_dup[edu_dropped_dup['high_income'] == False]['gross_tertiary_education_enrollment']
mean_low_tertiary = low_tertiary.mean()
print(mean_low_tertiary, '\n')

t_statistic, p_value = ttest_ind(high_tertiary, low_tertiary)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

The t-test indicates that there is a significant difference in gross tertiary enrollment between high-income and low-income countries, with higher tertiary education enrollment in high-income countries. This aligns with the chart, showing that high-income countries are more likely to have people enrolled in higher education, likely due to differences in opportunities available.

# Discusison and Conclusion

Our study tested several hypotheses concerning the distribution and characteristics of billionaires in high-income versus low-income countries, focusing on their industry involvement, gender, and age. We hypothesized that in developing countries, the majority of billionaires would emerge from primary and secondary industries, while in developed countries, a higher proportion of billionaires would be found within the tertiary sector. Additionally, we hypothesized that globally, regardless of the country's economic status, the majority of billionaires would be male and older in age.

Nevertheless, it is clear that the majority of billionaires in both types of countries is male. Our median is 65 years old which is considered the age of a senior citizen. Thus, it can be said that the majority of billionaires in our dataset do tend to be older in age.

Now we will look at the types of industries the billionaires in our dataset made their living in. The data analysis mostly confirmed our hypotheses to a significant extent. We observed that manufacturing, a secondary industry, was the most prominent industry among billionaires in developing countries. This reflects the economic structures of these countries, where manufacturing, agriculture, and extraction industries often serve as the backbone of the economy. Although, we also saw that some of the top industries present in low-income countries fell into the tertiary category so we must keep in mind that it is not uncommon for billionaires from lower-income countries to be involved in tertiary industries as well. In developed countries, our findings strongly align with our hypothesis that the tertiary sector–mostly finance, technology, and fashion/retail–plays a pivotal role in wealth accumulation. This trend underscores the shift towards knowledge-based economies in these regions, where innovation, financial services, and consumer markets drive economic growth. 

From our data, we also found that most billionaires are from a few specific countries. For instance, 754 of 2640 billionaires are from the United States, and 523 are from China. In addition, we looked into the relationship between the tax rate of a country and the number of billionaires in the country. According to our t-test, there is no statistically significant evidence to suggest that the tax rate of countries has an impact on the number of billionaires.

Since a majority of billionaires are from high-income countries, this may have impacted their ability to enroll in higher education, resulting in better or more options for their financial success.

Regarding the demographic characteristics of billionaires, our analysis substantiated the hypothesis that the majority are men and tend to be older. This finding points to persistent gender disparities in access to capital and opportunities for wealth generation, reflecting broader societal norms and economic barriers that women face. This is exacerbated even more so in developing economies.
